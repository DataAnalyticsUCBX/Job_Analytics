```python
#Imports dependencies
import pandas as pd
import os

url = "US-based jobs.csv"

```

```python
#Filter columns, eliminating Geo, Url, and Last date to apply. Leaving only Company name, job Title, Job Text, Job Location and Job PostDate
original_df=pd.read_csv(url)
original_df = original_df[["JobTitle","company_name","JobText","JobLocation","JobPostDate"]]
original_df.head()
```

```python
#Renaming columns to Position, Company, Description and reorganizing
original_df["Position"]=original_df["JobTitle"]
original_df["Company"]=original_df["company_name"]
original_df["Description"]=original_df["JobText"]
original_df=original_df[["Position","Company","Description","JobLocation","JobPostDate"]]
original_df.count()
```

```python
#Filter out the results we're interested with: 

Keywords = ["Data Scien", "Data Analy", "Data Eng", "Business Intelligence", "Tableau", "VBA", "Python", "JSON", "NumPy",  "Pandas", "Matplotlib", "Beautiful Soap", "Statistics Modeling", "API", "SQL", "MongoDB", "ETL", "HTML", "CSS", "JavaScript", "AJAX", "D3", "Leaflet", "Hadoop", "Machine Learning", "Product Analyst", "BI Analyst"] 

 

filteredUSjobs_df = original_df[original_df["Position"].str.contains("|".join(Keywords))] 
filteredUSjobs_df.count()
```
