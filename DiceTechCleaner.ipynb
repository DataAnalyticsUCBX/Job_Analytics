```python
import pandas as pd
job_data_to_load = "dice_us_tech_original.csv"
job_data = pd.read_csv(job_data_to_load)
```

```python
# drop unneeded columns
del job_data["advertiserurl"]
del job_data["employmenttype_jobstatus"]
del job_data["jobid"]
del job_data["postdate"]
del job_data["shift"]
del job_data["uniq_id"]

# # split joblocation_address into city and state
new=job_data["joblocation_address"].str.split(",",n=1,expand=True)
job_data["Loc_city"]=new[0]
job_data["Loc_state"]=new[1]
del job_data["joblocation_address"]
# drop rows where required fields are empty
job_data = job_data.loc[job_data.Loc_state.notnull()]
job_data = job_data.loc[job_data.company.notnull()]
# Remove listings for Canadian states
job_data = job_data.loc[~job_data.Loc_state.str.contains("ON")]
job_data = job_data.loc[~job_data.Loc_state.str.contains("BC")]

# replace all links in site_name column with dice.com
job_data.loc[:,"site_name"] = "dice.com"
# # rename columns to prep for merging with other datasets
job_data = job_data.rename(columns={"company":"Company","jobdescription":"Description","jobtitle":"Position","site_name":"Source"})
job_data.head(10)
```
